```python
def stringReplace(str):
    correct_mapping = {
        "\n": "",
        " ship ": " vận chuyển ",
        " Ship ": " vận chuyển ",
        "shop": "cửa hàng",
        "Shop": "cửa hàng",
        " m ": " mình ",
        " M ": " mình ",
        " mik ": " mình ",
        " Mik ": " mình ",
        " k ": " không ",
        " K ": " không ",
        " kh ": " khách hàng ",
        " KH ": " khách hàng ",
        " tl ": " trả lời ",
        " TL ": " trả lời ",
        " r ": " rồi ",
        " fb ": " mạng xã hội ", # facebook
        " FB ": " mạng xã hội ", # facebook
        " face ": " mạng xã hội ",
        " Tks ": " cảm ơn ",
        " Thanks ": " cảm ơn ",
        " thank ": "cảm ơn",
        " Thank ": "cảm ơn",
        " tks ": " cảm ơn ",
        " Tks ": " cảm ơn ",
        " tk ": " cảm ơn ",
        " Tk ": " cảm ơn ",
        " TK ": " cảm ơn ",
        "cam on": "cảm ơn",
        "Cam on": "cảm ơn",
        "sp": "sản phẩm",
        "Sp": "sản phẩm",
        "SP": "sản phẩm",
    }
    str = str.strip()
    for oldStr, replaceStr in correct_mapping.items():
        str = str.replace(oldStr, replaceStr)
    return str
```

```python
train_data = pd.read_csv("data/train.crash", sep="\t", header=None)
print(train_data.info())
```

```python
i = 0
lenght = train_data.shape[0]
dataset = pd.DataFrame(columns=["id", "text", "label"])
for i in range(0,lenght,3):
    text = stringReplace(train_data.iloc[i + 1, 0])
    row = {'id': train_data.iloc[i, 0], "text": text, "label": train_data.iloc[i + 2,0]}
    #print(row)
    dataset = dataset.append(row, ignore_index=True)
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer()
#count.fit(dataset.text)
```

```python
words = count.fit_transform(dataset.text)
count.get_feature_names()
```

```python
words.toarray()
```

```python
x = words.toarray()
y = dataset.label
# y = y.replace({"ham": 1, "spam": 0})
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_test_pred = model.predict(x_test)
y_train_pred = model.predict(x_train)
```

```python
# print("The accuracy score:", accuracy_score(y_test, y_pred) * 100)
print("The accuracy of traning:", accuracy_score(y_train, y_train_pred) * 100)
print("The accuracy of testing:", accuracy_score(y_test, y_test_pred) * 100)
print("Score of Scikit Learning:", model.score(x_test, y_test) * 100)
```

```python
test_data = pd.read_csv("data/test.crash", sep="\t", header=None)
j = 0
lenght = test_data.shape[0]
test_dataset = pd.DataFrame(columns=["id", "label"])
for i in range(0,lenght,2):
    text = stringReplace(test_data.iloc[i + 1, 0])
    x_new = np.array([text])
    x_new = count.transform(x_new)
    row = {'id': test_data.iloc[i, 0], "label": model.predict(x_new)[0]}
    #print(row)
    test_dataset = test_dataset.append(row, ignore_index=True)
test_dataset.to_csv('submission.csv', index=False)
```
