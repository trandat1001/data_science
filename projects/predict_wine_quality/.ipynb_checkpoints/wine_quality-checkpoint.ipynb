```python
le = preprocessing.LabelEncoder()
source.color = le.fit_transform(source.color)
```

```python
source.head()
```

```python
# Scale all continous data into one metric, we use RobustScale because all continous data got outliers
robust_scaler = preprocessing.RobustScaler()
source[number_type] = robust_scaler.fit_transform(source[number_type])
```

```python
source.head()
```

```python
#pf = preprocessing.PolynomialFeatures(degree=2)
#source = pf.fit_transform(source)
#source
```

```python
source.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(source, target, test_size=0.3)
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)
```

```python
print("R2 Score of traninning:", r2_score(y_train, y_train_pred) * 100)
print("Score of traninning:", lr.score(x_train, y_train) * 100)
print("Score of testing:", r2_score(y_test, y_test_pred) * 100)
print("MSE of testing:", mean_squared_error(y_test, y_test_pred))
```

```python
target_hat = lr.predict(source)
sb.distplot(target, label="Actual value", hist = False)
sb.distplot(target_hat, label="Predict value", hist = False)
```

```python
report = pd.DataFrame({"Algorithm": ["Multiple LinearRegression"], 
                       "R2 Score": [r2_score(y_test, y_test_pred) * 100]
                       })
report
```
