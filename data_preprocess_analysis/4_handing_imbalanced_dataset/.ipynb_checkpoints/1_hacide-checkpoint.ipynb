### Do chưa có mô hình nên ta dự đoán theo cảm nhận, dữ liệu type 1 quá ít nên ta cần làm cân bằng lại 
### Ta áp dụng over-sampling để scale type 1 lên gần bằng type 0

```python
from imblearn.over_sampling import SMOTE
```

```python
x = dataset[['x1', 'x2']]
y = dataset[['cls']]
smote = SMOTE()
x_s, y_s = smote.fit_resample(x, y)
```

```python
new_dataset = pd.DataFrame(x_s, columns=['x1', 'x2'])
new_dataset['cls'] = y_s
```

```python
new_dataset.head()
```

```python
new_dataset.cls.value_counts()
```

```python
plt.figure(figsize=(10, 10))
sb.scatterplot(x = new_dataset.x1, y = new_dataset.x2, hue=new_dataset.cls)
plt.show()
```
