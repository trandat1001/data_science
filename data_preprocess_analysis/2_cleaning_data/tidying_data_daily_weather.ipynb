### Day number is stored in columns. We need to convert them into on column 'day'
### There are many NaN value on day number

```python
new_dataset = pd.melt(frame = dataset, id_vars=["id", "year", "month", "element"], var_name="day")
```

```python
new_dataset.info()
```

```python
new_dataset.day = new_dataset.day.apply(lambda r: r[1])
```

```python
new_dataset.head()
```

```python
new_dataset[['year', 'month', 'day']] = new_dataset[['year', 'month', 'day']].apply(lambda c: pd.)
```

```python
import datetime
def get_date(row):
    return datetime.datetime(year=int(row["year"]), month=int(row["month"]), day=int(row["day"]))

new_dataset['date'] = new_dataset.apply(lambda row: get_date(row), axis = 1)
new_dataset = new_dataset.drop(['year', 'month', 'day'], axis=1)
```

```python
new_dataset = new_dataset.dropna()
```
