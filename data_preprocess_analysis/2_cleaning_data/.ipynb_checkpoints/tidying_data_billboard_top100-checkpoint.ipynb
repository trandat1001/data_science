```python
dataset = pd.read_csv("data/billboard.csv", encoding="mac_latin2")
```

```python
dataset.head()
```

### The issue of this dataset
* The week values are in column
* There are many NULL values in week

```python
new_dataset = pd.melt(frame=dataset, id_vars=["year","artist.inverted", "track","time","genre", "date.entered", "date.peaked"], 
                      var_name="week", value_name="rank")
```

```python
new_dataset.head()
```

```python
new_dataset.week = new_dataset.week.str.extract(r'(\d)').astype(int)
new_dataset = new_dataset.dropna()
```

```python
new_dataset.info()
```

```python
new_dataset.head()
```

```python
song = new_dataset[["year","artist.inverted", "track", "time", "genre"]]
```

```python
song = song.drop_duplicates()
song.reset_index(drop = True)
song['id'] = song.index
```
