### Requirement

```python
student = pd.read_excel("data/student.xlsx")
student
```

```python
student.info()
```

### All subject values are in columns: Physics, Chemistry, English, Math. We need to convert them into one column call  Subjects

```python
new_student = pd.melt(frame = student, id_vars = "Student", 
                                      var_name = "Subject",
                                      value_vars = ["Physics", "Chemistry", "English", "Math"],
                                      value_name = "Score"
                                      )
```

```python
new_student
```

```python
student_pivot = new_student.pivot(index = "Student", columns="Subject", values="Score")
```
