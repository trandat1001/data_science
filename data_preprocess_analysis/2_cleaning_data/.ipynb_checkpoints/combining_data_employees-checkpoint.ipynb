### Câu 1+: Cho dữ liệu employees1.csv và employees2.csv
* Đọc dữ liệu từ 2 tập tin trên vào 2 DataFrame với index của các DataFrame là 'employee'
* Kết hợp 2 DataFrame trên thành 1 DataFrame dùng chung 1 index là 'employee' (gợi ý: dùng left_index hoặc/và right_index hoặc dùng dataframe1.join(dataframe2))

```python
new_employee1 = pd.read_csv("data/employees1.csv", index_col=0)
new_employee2 = pd.read_csv("data/employees2.csv", index_col=0)
new_employee1 = new_employee1.set_index("employee")
new_employee2 = new_employee2.set_index("employee")
```

```python
new_employee1
```

```python
new_employee2
```

```python
new_employee = pd.merge(left = new_employee1, right = new_employee2, left_index=True, right_index =  True)
```

```python
new_employee
```

### Câu 5: Cho dữ liệu như sau:


* df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
             'food': ['fish', 'beans', 'bread']},
            columns=['name', 'food']) <br/>
* df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
             'drink': ['wine', 'beer']},
            columns=['name', 'drink'])

* Kết hợp 2 bộ dữ liệu này với tham số how='inner', how='outer', how='left', how='right'. Quan sát kết quả trong từng trường hợp.

```python
df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],'food': ['fish', 'beans', 'bread']},columns=['name', 'food'])
df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],'drink': ['wine', 'beer']}, columns=['name', 'drink'])
```

```python
df6
```

```python
df7
```

```python
df_inner = pd.merge(left=df6, right=df7, how="inner")
df_left = pd.merge(left=df6, right=df7, how="left")
df_right = pd.merge(left=df6, right=df7, how="right")
df_outer = pd.merge(left=df6, right=df7, how="outer")
```

```python
df_inner
```

```python
df_left
```

```python
df_right
```

```python
df_outer
```

### Câu 6: Cho dữ liệu như sau:

* df8 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
             'rank': [1, 2, 3, 4]}) <br/>
* df9 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
             'rank': [3, 1, 4, 2]})

* Kết hợp 2 bộ dữ liệu này với tham số on='tên_cột_trùng_lắp' và/hoặc suffixes=["_L", "_R"]


```python
df8 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],'rank': [1, 2, 3, 4]})
df9 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],'rank': [3, 1, 4, 2]})
```

```python
df8
```

```python
df9
```

```python
df10 = pd.merge(left = df8, right = df9, on = "name")
```

```python
df10
```

```python
df11 = pd.merge(left = df8, right = df9, on = "name", suffixes=["_L", "_R"])
```

```python
df11
```
